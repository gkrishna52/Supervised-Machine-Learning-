```python
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
print("Data imported")
```

### Extracting Dependent And Independent variables from dataset.
#### Hours is independent Variable and Scores is dependent Variable.

```python
x= data.iloc[:,:-1].values
y = data.iloc[:,1].values
```

#### In the above lines of code, for x variable, I have taken -1 value since I want to remove the last column from the dataset. For y variable, I have taken 1 value as a parameter, since i want to extract the second column and indexing starts from the zero

```python
x
```

```python
y
```

###  I will split both variables into the test set and training set. I have 25 observations, so i will take 15 observations for the training set and 10 observations for the test set. I am splitting our dataset so that we can train our model using a training dataset and then test the model using a test dataset using Sklearn.

```python
from sklearn.model_selection import train_test_split
X_train, x_test,Y_train,y_test = train_test_split(x,y, test_size = 1/2.5, random_state=0) 
```

### Fitting to Training Set:

#### I will import the LinearRegression class of the linear_model library from the scikit learn. After importing the class, I am going to create an object of the class named as a regressor

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,Y_train)

print("Training complete.")
```

### Prediction of test set result:

```python
y_pred = regressor.predict(x_test)
x_pred = regressor.predict(X_train)
```

### Comparing Actual VS Predicted

```python
df1 = pd.DataFrame({'Actual': y_test, 'Predcited': y_pred})
df1
```

## Visualizing the Training set results

```python
mtp.scatter(X_train, Y_train, color="green")   
mtp.plot(X_train, x_pred, color="red")    
mtp.title("Hours vs Percentage (Training Dataset)")  
mtp.xlabel("Hours Studied(hr)")  
mtp.ylabel("Percentage Score(%)")  
mtp.show()   
```

## Visualizing the Test set results 

```python
mtp.scatter(x_test, y_test, color ="blue")
mtp.plot(X_train, x_pred, color ="black")
mtp.title("Hours vs Percentage (Training Dataset)")  
mtp.xlabel("Hours Studied(hr)")  
mtp.ylabel("Percentage Score(%)")  
mtp.show()   
```

### Testing Own data

```python
Hours = [[9.25]]
own_pred = regressor.predict(Hours)
print("No of Hours = {}".format(Hours))
print("Predicted Score = {}".format(own_pred[0]))
```

## Evaluating Model

```python
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred)) 
```
